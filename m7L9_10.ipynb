```python
years = np.array(['2012', '2013', '2014', '2015'])

sales_africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])

sales_USCA = np.array([492756.60, 486629.30, 627634.98, 757108.13])

sales_LATAM = np.array([385098.15, 464733.29, 608140.77, 706632.93])

sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])

sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
```

```python
# Creating the plot
fig, ax = plt.subplots()

# Plotting each region with labels
europe, = ax.plot(years, sales_Europe, label="Europe")
europe.set_dashes([2,2,2,2])
usca, = ax.plot(years, sales_USCA, label="USCA")
africa, = ax.plot(years, sales_africa, label="Africa")
asia, = ax.plot(years, sales_Asia_Pacific, label="Asia Pacific")
latam, = ax.plot(years, sales_LATAM, label="LATAM", )
latam.set_dashes([2,5,5,2])

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating a 2x3 grid of subplots with shared x and y axes
fig, ax = plt.subplots(ncols=3, nrows=2, sharex=True, sharey=True, figsize=(12, 8))

# Plotting each region with labels on individual subplots
europe, = ax[0, 0].plot(years, sales_Europe, label="Europe")
europe.set_dashes([2, 2, 2, 2])  # Custom dash pattern

usca, = ax[0, 1].plot(years, sales_USCA, label="USCA")

africa, = ax[0, 2].plot(years, sales_africa, label="Africa")

asia, = ax[1, 0].plot(years, sales_Asia_Pacific, label="Asia Pacific")

latam, = ax[1, 1].plot(years, sales_LATAM, label="LATAM")
latam.set_dashes([2, 5, 5, 2])  # Custom dash pattern for LATAM

# Adding legends to each subplot
ax[0, 0].legend()
ax[0, 1].legend()
ax[0, 2].legend()
ax[1, 0].legend()
ax[1, 1].legend()

# Removing the unused subplot (bottom-right corner)
fig.delaxes(ax[1, 2])

# Showing the plot
plt.show()
```
