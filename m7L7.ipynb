```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
pd_category = np.array(['Furniture','Technology','office Supply'])
sales = np.array([4110059.49,4745557.60,3787492.84])

plt.bar(pd_category,sales)
```

```python

plt.bar(pd_category, sales, width=0.5, color='brown')
plt.title("Sales across product categories\n", fontdict={'fontsize':16, 'fontweight':5, 'color':'green'})
plt.xlabel("\ncategories", fontdict={'fontsize':13, 'fontweight':5, 'color':'blue'})
plt.ylabel("\nsales", fontdict={'fontsize':13, 'fontweight':5, 'color':'blue'})

# Adding custom y-ticks
ticks = np.arange(0, 6000000, 1000000)  
labels = ["{}M".format(i // 1000000) for i in ticks]  
plt.yticks(ticks, labels)

plt.show()
```
